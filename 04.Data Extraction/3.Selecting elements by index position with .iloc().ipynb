```python
#read csv file
df=pd.read_csv('players_20.csv')

#set index
df.set_index('short_name',inplace=True)

# select column
df=df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
```

# iloc()

.iloc() is used for integer-location based indexing—that means you select rows and columns by their numerical position (0-based index), not by labels.

🔹 Basic Syntax

    df.iloc[row_index, column_index]

🔹 Key Points

    Uses integer positions (0, 1, 2, …)
    Works like standard Python indexing
    Supports slicing, lists, and boolean arrays

## 1.Selecting with a single value

```python
#get the height of L.Messi

row=df.index.get_loc('L. Messi')
column=df.columns.get_loc('height_cm')

df.iloc[row,column]
```

```python
# get the weight of Cristiano Ronaldo

row=df.index.get_loc('Cristiano Ronaldo')
column=df.columns.get_loc('weight_kg')

df.iloc[row,column]
```

```python
# get all rows inside the 'height_cm' column

column=df.columns.get_loc('height_cm')

df.iloc[:,column]
```

```python
# get all columns that corresponds to the index 'L. Messi'

row=df.index.get_loc('L. Messi')

df.iloc[row,:]
```

## 2.Selecting with a list of values

```python
# get all data about L. Messi and Cristiano Ronaldo

row1=df.index.get_loc('L. Messi')
row2=df.index.get_loc('Cristiano Ronaldo')

df.iloc[row1:row2+1,:]
```

```python
# get the height of L. Messi and Cristiano Ronaldo

row1=df.index.get_loc('L. Messi')
row2=df.index.get_loc('Cristiano Ronaldo')
column=df.columns.get_loc('height_cm')

df.iloc[row1:row2+1,column]
```

```python
# get the height and weight of L. Messi

row=df.index.get_loc('L. Messi')
column1=df.columns.get_loc('height_cm')
column2=df.columns.get_loc('weight_kg')

df.iloc[row,column1:column2+1]
```

```python
# get the height and weight of L. Messi and Cristiano Ronaldo

row1=df.index.get_loc('L. Messi')
row2=df.index.get_loc('Cristiano Ronaldo')
column1=df.columns.get_loc('height_cm')
column2=df.columns.get_loc('weight_kg')

df.iloc[row1:row2+1,column1:column2+1]
```

## 3.Selecting a range of data with a slice

start:stop:step

HERE THE STOP IS NOT INCLUDED

```python
# slice column labels age/club
#players=['L. Messi','Cristiano Ronaldo']

player=[0,1]

df.iloc[player,1:7]
```

```python
# slice index labels (top1 to top10 player)
# columns=['age','dob','height_cm','weight_kg']

column_index=df.columns.get_indexer(['age','dob','height_cm','weight_kg'])
df.iloc[0:10,column_index]
```

## 4.Selecting with conditions

iloc **cannot accept a **boolean Series****,but only a **BOOLEAN LIST**.
We have to use the list() funciton to convert a Series into a boolean list.

```python
#one condition: Select player with the height above 180cm
# columns=['age','dob','height_cm','weight_kg']

columns_index=df.columns.get_indexer(['age','dob','height_cm','weight_kg'])

player_list=list(df['height_cm']>180)

df.iloc[player_list,columns_index]
```

```python
# multiple conditions: select player with height above 180 cm and nationaly is argentina

player_list=list((df['height_cm']>180) & (df['nationality']=='Argentina'))
columns_index=df.columns.get_indexer(['age','dob','height_cm','weight_kg'])

df.iloc[player_list,columns_index]
```

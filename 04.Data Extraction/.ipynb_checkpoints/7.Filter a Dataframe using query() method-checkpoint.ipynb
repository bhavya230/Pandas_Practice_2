```python
import pandas as pd
```

```python
# reads csv file
df = pd.read_csv('players_20.csv')

# set index
df.set_index('short_name', inplace=True)
# select column

df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df.head()
```

# query()
.query() is used to filter rows of a DataFrame using a condition written as a string.

👉 It’s an alternative to:

df[df['age'] > 18]

🔹Basic Syntax:

df.query("condition")

🔴 Important Notes:

You must pass condition as a string.

Column names are used directly (no df['col']).

Use @ for external variables

✅ Simple condition: 
df.query("age > 18")

✅ Multiple conditions: 
df.query("age > 18 and salary > 50000")

✅ Using OR: 
df.query("age > 18 or salary > 50000")

✅ Using variables (important): 

limit = 18

df.query("age > @limit")

✅ Column names with spaces: 
df.query(" `total salary` > 50000")

👉 Use backticks `

```python
# select players older than 34
df.query(" age > 34 ").head()

# exercise: write the equivalent boolean indexing
df[df['age']>34].head()
```

```python
# select player older than 34 and from Italy
df.query(" age>34 and nationality=='Italy' ").head()

# using boolean indexing
df[(df['age']>34) & (df['nationality']=='Italy')].head()
```

```python
# add a not operator to first example
df.query("not (age>34 and nationality=='Italy') ").head()

# boolean
df[~((df['age']>34) & (df['nationality']=='Italy'))].head()
```

```python
# convert height to meters and select those with heihgt above 1.8

df['height_m']=df['height_cm'] / 100

df.query(" height_m >1.8").head()

# boolean indexing/slicing
df[df['height_m']>1.8].head()
```

```python
# select players that were born after 1990

#check datatype of dob
df.dtypes # or use df.info()

# convert 'dob' column  to datetime format.
df['dob']=df['dob'].astype('datetime64[ns]')
df['dob'].dt.year

#query
df.query(" dob.dt.year > 1990 ")

# boolean indexing
df[df['dob'].dt.year >1990]
```

## Multiple leagues
https://www.football-data.co.uk/mmz4281/2526/E0.csv

https://www.football-data.co.uk/mmz4281/2526/E1.csv

https://www.football-data.co.uk/mmz4281/2526/E3.csv

https://www.football-data.co.uk/mmz4281/2526/EC.csv

```python
#creating list of leagues
leagues=['E0','E1','E2','E3','EC']
frames=[] # to store all the dataframes

#Looping through leagues and read multiple csv and append it into list

for league in leagues:
    df=pd.read_csv(root+"2526"+"/"+league+".csv")
    frames.append(df)
```

```python
#length of frame
len(frames)
```

```python
# show frames
for data in frames:
    print(data.head())
```

# Multiple seasons

```python
for season in range(15,21):
    print(str(season)+str(season+1))
```

```python
#creating list of leagues
leagues=['E0','E1','E2']
frames=[]

#loop through range of seasons and insert new column 
for seasons in range(15,21): # 6 seasons
    for league in leagues: #3 leagues
        df=pd.read_csv(root+str(season)+str(season+1)+"/"+league+".csv",encoding="unicode_escape")
        df.insert(1,'season',season)
        frames.append(df)
```

```python
# toatl frames (6seasons X 3 leagues)
len(frames)
```

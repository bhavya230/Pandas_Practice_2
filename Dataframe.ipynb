```python
#importing libraries
import pandas as pd
import numpy as np
```

# 1.Creating a DATAFRAME from an array

```python
# creating the array
arr=np.array([[1,2],[2,3],[4,5]])
arr
```

```python
#creating a datafarme
pd.DataFrame(arr)
```

```python
# creating DataFrame with customized row(index) and columns name
df=pd.DataFrame(arr,index=['row1','row2','row3'],columns=['col1','col2'])
df
```

```python
import pandas as pd
```

```python
# reading the csv file
df=pd.read_csv("StudentsPerformance.csv")
df.head()
```

# 1 Value_Counts
Counts no of diff categories entries for the categorial feature in the dataset.

```python
# counting gender elements

# len() function
len(df['gender'])

# .count() method
df['gender'].count()
```

```python
# counting gender elements by category
df['gender'].value_counts()
```

```python
#return the relative frequency (No of times an event occur/Total no of Observations)
df['gender'].value_counts(normalize=True)
```

```python
# counting "parental level of education" elements by category
df['parental level of education'].value_counts()
```

```python
# return the relative frequency and round to 2 decimals
df['parental level of education'].value_counts(normalize=True).round(2)
```

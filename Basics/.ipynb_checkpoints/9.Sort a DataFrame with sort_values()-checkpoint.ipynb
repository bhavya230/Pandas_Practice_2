```python
import pandas as pd
```

```python
#reading and showing the DataFrame
df=pd.read_csv("StudentsPerformance.csv")
df.head()
```

# 1. Sort a DataFrame

```python
# sort by one column
df.sort_values(by='math score')
```

```python
# sort descending by 1 column
df.sort_values('math score',ascending=False)
```

```python
# sort descending by multiple columns
df.sort_values(['math score','reading score'],ascending=False)
```

```python
# change is not reflected to the original DataFrame : Only copy is created
df.head()
```

```python
# sort descending by multiple columns and update DataFrame
df.sort_values(['math score','reading score'],ascending=False,inplace=True)
```

```python
# show the dataframe
df.head() # dataframe updated
```

```python
# sort descending with a key function
df.sort_values('race/ethnicity',key=lambda col: col.str.lower() )
```

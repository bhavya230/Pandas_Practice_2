# copy()

the .copy() function is used to create a copy of a DataFrame or Series.

**📌 Why .copy() is important**

Without .copy(), sometimes variables may share the same data, so changing one can unintentionally change the other.

**📌 Syntax**

df_copy = df.copy(deep=True)

**📌 Without .copy() (problem case)**

df2 = df   # no copy, just reference

df2['A'][0] = 100

print(df)   # also changes!

👉 Both change because they point to the same object

## 1 DEEP COPY
df_copy = df.copy(deep=True)

Completely independent copy

Changes do NOT affect original

```python
df_copy=df.copy()
df_copy.head()
```

```python
# update value in original dataframe

df['height_cm']=170
df.head()
```

```python
# copy vs original dataframe
df_copy.head()

df.head()
```

## 2 SHALLOW COPY
df_copy = df.copy(deep=False)

Shares underlying data

Changes may affect both

```python
df_copy=df.copy(deep=False)
df_copy.head()
```

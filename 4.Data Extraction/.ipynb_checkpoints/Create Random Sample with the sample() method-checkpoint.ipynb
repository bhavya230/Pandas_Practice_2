# sample()

🔹 What is .sample()?

.sample() is used to randomly select data from a DataFrame or Series.

By default → it selects random rows.
You can also select random columns

🔹 Basic Syntax:

**df.sample(n=None, frac=None, axis=0, replace=False, random_state=None).**

🧩 Parameters Explained

**1. n (Number of items):**
Specifies exact number of rows/columns to return

**2. frac (Fraction of data):**
Returns a percentage of the dataset


df.sample(frac=0.5)

👉 Returns 50% of the rows

✔ You use either n OR frac, not both

**3. axis**

Controls whether sampling is from rows or columns

Value	    Meaning

0 (default)	 Sample rows

1	         Sample columns

**4. replace:**
Controls whether sampling is done with replacement.

Value	        Meaning

False (default)	No repetition

True	        Allows duplicates

**5. random_state:**
Sets a fixed seed for reproducibility.

df.sample(n=3, random_state=42)

👉 You’ll get the same random result every time

```python
# extract 10 random elements from the "nationality" column
df['nationality'].sample(n=10) # every time diff result

df['nationality'].sample(n=10,random_state=42) # same result everytime
```

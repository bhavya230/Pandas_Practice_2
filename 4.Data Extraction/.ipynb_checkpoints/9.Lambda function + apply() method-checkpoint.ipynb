```python
import pandas as pd
```

```python
# reads csv file
df = pd.read_csv('players_20.csv')

# set index
df.set_index('short_name', inplace=True)
# select column

df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

# 1. Lambda function

lambda functions are small, anonymous (unnamed) functions used for short, one-time operations.A lambda function is a one-line function defined using the keyword lambda instead of def.

**📌 Syntax:**

lambda arguments: expression

👉 It can have:

Any number of arguments.

Only one expression (no multiple statements)

```python
# basic function
def sum_values(a,b):
    return a + b

sum_values(2,3)
```

```python
sum_values
```

```python
# lambda function

sum2= lambda a,b: a+b

sum2(2,3)
```

```python
sum2
```

# 2 .apply() + Lambda function

```python
# use lambda function to convert "height_cm" series to meters
df.head()

df['height_cm'].apply(lambda height:height/100)

# alternative
df['height_cm']/100
```

```python
# use lambda function to convert "long_name" seris to uppercase
df['long_name'].apply(lambda name: name.upper())

# alternative
df['long_name'].str.upper()
```

```python
# use lambda function to get the year or "dob" series

df['dob']=pd.to_datetime(df['dob'])

df['dob'].apply(lambda x: x.year)

# alternative with dt
df['dob'].dt.year
```

```python
# apply lambda function to dataframe in order to calculate bmi

df.apply(lambda x :x['weight_kg']/pow(x['height_cm']/100,2),axis=1)
```

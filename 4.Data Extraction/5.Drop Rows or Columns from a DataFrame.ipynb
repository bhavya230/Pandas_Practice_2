```python
import pandas as pd
```

```python
# read csv file
df=pd.read_csv('players_20.csv')

# set index
df.set_index('short_name',inplace=True)
df
#set column
df=df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]	
```

```python
# show dataframe
df.head()
```

# .drop()

the .drop() method is very flexible—it can remove rows or columns depending on the parameters you pass

df.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')

1. labels


   Specifies what to delete

   
    Can be:

   
        Single label → 2

        List → [1, 3, 5]

    👉 Works for both rows and columns depending on axis

3. axis → alternative way

    axis=0 → rows
    axis=1 → columns

3.index → drop rows

4.columns → drop columns

5.inplace=True → change original DataFrame

6.errors='ignore' → avoid error if not found

 ## 1.Drop row

```python
#drop 1 row
#axis paramter

df.drop(labels='L. Messi',axis=0)

#or index parameter
df.drop(index='L. Messi')
```

```python
# drop two or more rows and update the data
df.drop(index=['Pan Ximing','Wang Haijian','Zhang Wei'],inplace=True)
```

```python
# show dataframe
df
```

## 2 Drop Columns

```python
#drop 1 column
#axis parameter
df.drop(labels='club',axis=1)

#columns paramter
df.drop(columns='club')
```

```python
# drop columns by position/index (drop last column)
df.columns[[-1]]

df.drop(columns=df.columns[[-1]])
```

```python
#drop 2 pr more columns and update the data

df.drop(columns=['weight_kg','nationality'],inplace=True)
```

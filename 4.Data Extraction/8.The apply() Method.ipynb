# apply()

.apply() is a powerful and flexible method used to apply a function across a Series or DataFrame..apply() lets you run a function on each element, row, or column of your data.

**📌 Syntax:**

    👉 On Series (single column):
    
        series.apply(function)

    👉 On DataFrame:
    
        df.apply(function, axis=0)  # column-wise (default)
        
        df.apply(function, axis=1)  # row-wise

```python
# use numpy function abd apply it to series

import numpy as np

df['age'].apply(np.sqrt)
```

```python
# use your own function and apply it to the dataframe

def calculate_bmi(row):
    return row['weight_kg']/pow(row['height_cm']/100,2)
```

```python
df.apply(calculate_bmi,axis=1) #function receives each row as a Series
```

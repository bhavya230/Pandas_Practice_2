# 1. Lambda function

lambda functions are small, anonymous (unnamed) functions used for short, one-time operations.A lambda function is a one-line function defined using the keyword lambda instead of def.

**📌 Syntax:**

lambda arguments: expression

👉 It can have:

Any number of arguments.

Only one expression (no multiple statements)

```python
# basic function
def sum_values(a,b):
    return a + b

sum_values(2,3)
```

```python
sum_values
```

```python
# lambda function

sum2= lambda a,b: a+b

sum2(2,3)
```

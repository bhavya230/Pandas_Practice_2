# .loc()
the loc[] method is used for label-based indexing — meaning you select data using row and column names (labels), not positions.

🔹 📌 Basic Idea

Access rows/columns using labels

Works with row index names and column names

Supports filtering, slicing, and assignment

🔹 ✅ Syntax

df.loc[row_labels, column_labels]

```python
import pandas as pd
```

```python
#read csv file
df=pd.read_csv('players_20.csv')
#set index
df.set_index('short_name',inplace=True)
# select column
df=df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
```

# .loc()
the loc[] method is used for label-based indexing — meaning you select data using row and column names (labels), not positions.

🔹 📌 Basic Idea

Access rows/columns using labels

Works with row index names and column names

Supports filtering, slicing, and assignment

🔹 ✅ Syntax

df.loc[row_labels, column_labels]

## 1. Selecting with a single value

loc[row_label,column_label]

```python
#get all data about L.Messi

df.loc['L. Messi']
```

```python
# get the height of l.Messi

df.loc['L. Messi','height_cm']
```

```python
# get the weight of Cristiano Ronaldo

df.loc['Cristiano Ronaldo','weight_kg']
```

```python
# get all rows inside the 'height_cm' column

df.loc[:,'height_cm']
```

```python
# get all columns that correspond to index 'L. Messi'

df.loc['L. Messi',:]
```

## 2 Selecting with a list of values

```python
# get all data about L.Messi and Cristiano Ronaldo
df.loc[['L. Messi','Cristiano Ronaldo'],:]
```

```python
# get the height of L. Messi,'Cristiano Ronaldo

df.loc[['L. Messi','Cristiano Ronaldo'],'height_cm']
```

```python
# get height and weight of L. Messi

df.loc['L. Messi',['height_cm','weight_kg']]
```

```python
# get height and weight of L. Messi and Cristiano Ronaldo

df.loc[['L. Messi','Cristiano Ronaldo'],['height_cm','weight_kg']]
```

## 3 Selecting a range of data with a slice

start:end:stop

**Here the end value is also inlcuded**

```python
# slice column labels

players=['L. Messi','Cristiano Ronaldo']

df.loc[players,'long_name':'club':2]
```

```python
# slice index labels
columns=['long_name','age','dob','height_cm','weight_kg','nationality','club']

# get top1 and top10 player name
df.index[:10]
df.loc['L. Messi':'M. Salah',columns]
```

# 1 unique()

In pandas, the unique() method is used to get distinct (unique) values from a column (Series).

**🔹 📌 Basic Idea**

Returns only unique values

Keeps the order of appearance

Works on Series (single column)

**✅ Syntax**

series.unique()

```python
# get unique elements in the "company" column
df['Company'].unique()
```

```python
# get unique elements in the "Inches" column
df['Inches'].unique()
```

```python
# get the length of the unique elemnts
len(df['Company'].unique())
```

# nunique()

In pandas, the nunique() method is used to count the number of unique (distinct) values.

**🔹 📌 Basic Idea**

    Returns a count, not the values

    Useful for understanding cardinality (how many distinct values)

**🔹 ✅ Syntax**

    series.nunique(dropna=True)
    
**Parameters:**

    dropna=True (default) → ignores missing values (NaN)
    dropna=False → counts NaN as a unique value

```python
# get no of unique elements in the company column
df['Company'].nunique()
```

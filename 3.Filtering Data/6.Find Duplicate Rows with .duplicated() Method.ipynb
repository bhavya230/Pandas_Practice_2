# 1 duplicated()

the duplicated() method is used to detect duplicate rows in a DataFrame (or duplicate values in a Series).

🔹 📌 Basic Idea

It returns a Boolean Series:

True → duplicate value

False → first occurrence (or unique)

🔹 ✅ Syntax

df.duplicated(subset=None, keep='first')

Parameters:

subset → column(s) to check duplicates on

keep → which duplicates to mark:

'first' (default): mark duplicates except first

'last': mark duplicates except last

False: mark all duplicates


### 1.1 Example

```python
# find duplicates in 1 column/series
df.duplicated('laptop_ID')
```

```python
# show elements in dataframe with duplicated in column 'laptop_ID'
df[df.duplicated('laptop_ID')]
```

#### 🔹 📌 How duplicated() works with multiple columns

When you pass 2 or more columns using the subset parameter, pandas checks row-wise combinations of those columns.

👉 A row is considered duplicate only if ALL specified column values match a previous row.

🔹 ✅ Syntax

df.duplicated(subset=['col1', 'col2'], keep='first')

```python
# DUPLICATES IN 2 OR MORE ROWS
df.duplicated(subset=['TypeName','Product','Inches'])
```

```python
# show all the values duplicated(except for the first) - this is the default behaviour
duplicated=df.duplicated(subset=['TypeName','Product','Inches'],keep='first')
duplicated
```

```python
df[duplicated]
```

```python
# sort the values
df[duplicated].sort_values(by=['Product','TypeName'])
```

#### 1.2 Example 2

Task:Find the cheapest and most expensive laptop of each company using the sort_values() and duplicate() methods(keep "first" , "last" and False)

```python
#sort the dataframe ascending by company and price
# cheapest first and most expensive last

df=df.sort_values(by=['Company','Price_euros'])
df
```

```python
# check all value category in "Company" column
df.value_counts('Company')
```

```python
# duplicated values in "Company" columns( keep='first')
duplicated_first=df.duplicated(subset='Company',keep='first')
duplicated_first
```

```python
# show dataframe with duplicated values in Company column
df[duplicated_first]

#keep='first' (cheapest laptop per company)
#show dataframe with non duplicated values in "company column" (those with false)
df[~duplicated_first][['Company','Price_euros']]

# check all value categories
df[~duplicated_first].value_counts('Company')
```

```python
df[duplicated_first].value_counts('Company')
```

```python
# keep='last'
duplicated_last=df.duplicated(subset='Company',keep='last')
duplicated_last
#keep=False
duplicated_false=df.duplicated(subset='Company',keep=False)
duplicated_false
```

```python
#keep='last' (most expensive laptop
#show dataframe with non duplicated values in "company" column
df[~duplicated_last][['Company','Price_euros']]
```

```python
#keep=Fasle( mark all as duplicated)
#show dataframe with non duplicated values in "Company" column
df[~duplicated_false][['Company','Price_euros']]
```

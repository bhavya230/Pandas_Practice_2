# Create a column based on 1 conditions: np.where()

The np.where() function in NumPy is used to find elements in an array that satisfy a condition — or to choose values based on a condition.

🔹 Basic Syntax

**np.where(condition, x, y)**

🔹 How it works

**condition: A boolean expression (e.g., arr > 5)**

**x: Value(s) to use where condition is True**

**y: Value(s) to use where condition is False**

🧠 Key Notes

Works element-wise on arrays

Returns:

Values array (if x and y provided)

Indices tuple (if only condition provided)

Very useful for filtering, masking, and conditional replacement

```python
import numpy as np
```

```python
# create an array based on price tiers
# price>2000 - Expensive ; price<2000 - Cheap
price_tiers=np.where(df['Price_euros']>2000,'Expensive','Cheap')
price_tiers
```

```python
# set it to a new column
df.insert(13,'Price_Tiers',price_tiers)
```

```python
# show dataframe
df.head()
```

```python
# count values in price_tier columns
df.value_counts('Price_Tiers')
```

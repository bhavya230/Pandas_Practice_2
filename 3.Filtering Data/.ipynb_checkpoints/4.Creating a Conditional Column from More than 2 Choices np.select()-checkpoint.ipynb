# 1 Creating a Conditional Column from More than 2 Choices : np.select()

The np.select method in NumPy is used to choose elements from different arrays (or values) based on multiple conditions. It’s especially useful when you have multiple conditions and want a clean alternative to nested if-else logic.

**🔹 Syntax**

**np.select(condlist, choicelist, default=0)**

**Parameters:**

condlist → list of boolean arrays (conditions)

choicelist → list of values/arrays to return when each condition is True

default → value if none of the conditions are True

```python
import numpy as np
```

#### create an array based on multiple price tiers (+2 choices)

To split laptop prices into “cheap, affordable, expensive” based on a meaningful “middle.” The best approach here is not just one middle value, but two cut points so you can form 3 categories.

The most practical and commonly used method is quartiles (percentiles) using NumPy.

✅ Best Approach: Use Percentiles (Q1, Median, Q3)

Instead of a single middle, do this:

Q1 (25%) → lower boundary

Q2 (50%) → median

Q3 (75%) → upper boundary

```python
# Decide values to use for divind the prices into 3 categories : cheap , affordable, expensive
Q1=np.percentile(df['Price_euros'],25) # 25% of the prices are below this value( could be used for cheap)
Q2=np.percentile(df['Price_euros'],50) # 50% pf prices are below this(affordable)
Q3=np.percentile(df['Price_euros'],75) #75% of the prices are below this(expensive)
print(Q1)
print(Q2)
print(Q3)
```

```python
# create "conditions" and "values" variable
condition=[df['Price_euros']<Q1,
          (df['Price_euros']>=Q1) & (df['Price_euros']<Q3),
           df['Price_euros']>=Q3
          ]

values=["Cheap","Affordable","Expensive"]
```

```python
#creating list of leagues
leagues=['E0','E1','E2']
frames=[]

#loop through range of seasons and insert new column 
for seasons in range(15,21): # 6 seasons
    for league in leagues: #3 leagues
        df=pd.read_csv(root+str(season)+str(season+1)+"/"+league+".csv",encoding="unicode_escape")
        df.insert(1,'season',season)
        frames.append(df)
```

# Organize All The Data in a Dictionary

```python
# creste s dictionary with original name of league as key
dict_countries={'Spanish La Liga':'SP1','Spanish Segunda Division':'SP2',
                'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
# get a dictionar y
dict_countries['Spanish La Liga']
```

```python
dict_historical_data={}

#creating list of leagues
#leagues=['E0','E1','E2']

#loop through key elements 
for league in dict_countries: # 3 seasons
    frames=[]
    for season in range(15,21): #6 leagues
        df=pd.read_csv(root+str(season)+str(season+1)+"/"+dict_countries[league]+".csv",encoding="unicode_escape")
        df.insert(1,'season',season)#insert new column
        frames.append(df)
    df_concat=pd.concat(frames) #concatenating all the seasons wise dataframes
    dict_historical_data[league]=df_concat # league(key) having all the seasons data as the value
```

```python
#show dataframe inside dict_historical_data
dict_historical_data['Spanish La Liga']
```

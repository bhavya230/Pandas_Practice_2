## 1.2 Adding a new column with an array
The DataFrame has 1000 elements, so we need to create an array with 1000 elemnts

```python
# create an array of 1000 elements
language_score=np.arange(0,1000,1)
type(language_score)
```

```python
#length of the array
len(language_score)
```

```python
# adding a new column to dataframe with an array
df['language score']=language_score
df
```

### Random numbers

```python
# Create Random Integer Numbers Between 1 and 100
int_language_score=(np.random.randint(low=1,high=100,size=1000))
```

```python
#min value inclusive and high value exclusive
print(min(int_language_score))
print(max(int_language_score))
```

```python
# adding a new column to dataframe with an array
df['language score']=int_language_score
df
```

# 1 Adding a new column

## 1.1 assign()
When to use assign ?

-Add multiple columns in a single line of code.

-When you need to overwrite the values of an existing columns(best practice).
    
**IMPORTANT** : It returns a new object(copy) withh all original columns in addition to the new ones. Follow Immutable style-> **original data safe**

```python
 import numpy as np
```

```python
#create random numbers for our new score columns
score1=np.random.randint(low=1,high=100,size=1000)
score2=np.random.randint(low=1,high=100,size=1000)
```

```python
#create the Series using the random numbers
series1=pd.Series(score1,index=np.arange(0,1000))
series2=pd.Series(score2,index=np.arange(0,1000))
```

```python
#using assign() to add multiple columns
df.assign(score1=series1,score2=series2)
```

```python
#assign method only creates a copy original remanins unaffected
df
```

```python
# to replicate the chnages in the orignal dataframe
df=df.assign(score1=series1,score2=series2)
```

```python
df
```

## 1.2 insert()
Inserts a new column at a specific position or index.
It does not return anything.
**It modifies the DataFrame directly (in-place)**
**Syntax: df.insert(loc,column,value)**

```python
#using index to add a coluumn at a specific position
df.insert(8,"language score",np.random.randint(low=1,high=100,size=1000))
```

```python
# show the dataframe
df
```

```python
import pandas as pd
```

```python
df=pd.read_csv('laptop_price.csv',encoding='unicode-escape')
```

```python
df.head()
```

# Filter dataframe based on a condition

```python
# find apple laptops
df[df['Company']=='Apple']
```

```python
# find laptops that costs more tha  2000 euros
df[df['Price_euros']>2000]
```

# 2 Filter dataframe based on multiple conditions

```python
# find apple laptops taht cost more tahn 2000 euros
(df['Company']=='Apple') & (df['Price_euros']>2000)
```

# 2.1  &(and)

```python
# filter dataframe based on multiple condition
df[(df['Company']=='Apple') & (df['Price_euros']>2000)]
```

# 2.2 |(OR)

```python
#find laptops Apple or Dell laptops
(df['Company']=='Apple') | (df['Company']=='Dell')
```

```python
# filter dataframe based on multpiple conditions
df[(df['Company']=='Apple') | (df['Company']=='Dell')].value_counts('Company')
```

```python
# find laptops from Apple Or Dell that costs over 2000 euros
df_DellOrApple=df[(df['Company']=='Apple') | (df['Company']=='Dell')]
df_PriceAbove2000=df[df['Price_euros']>2000]

df[((df['Company']=='Apple') | (df['Company']=='Dell')) & (df['Price_euros']>2000)]
```

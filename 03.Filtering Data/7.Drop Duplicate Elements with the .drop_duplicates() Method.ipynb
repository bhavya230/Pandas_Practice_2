# 1 drop_duplicates()

the drop_duplicates() method is used to remove duplicate rows from a DataFrame (or duplicate values from a Series).

🔹 📌 Basic Idea

Keeps only unique rows.
Removes repeated entries based on column values

🔹 ✅ Syntax

df.drop_duplicates(subset=None, keep='first', inplace=False)

Parameters:

subset → column(s) to check duplicates

keep → which duplicate to keep:

    'first' (default): keep first, remove rest
   
    'last': keep last, remove others
    
    False: remove all duplicates
    
    inplace → modify original DataFrame (True/False)

```python
# drop duplicates in 2 or more columns
df.drop_duplicates(subset=['Company'],keep='first')[['Company','Price_euros']]
```

```python
# sort dataframe ascending by company and price
#cheapest first and most expensive last
df_sorted=df.sort_values(['Company','Price_euros'])
```

```python
# cheapest first
df_cheapest=df_sorted.drop_duplicates(subset=['Company'],keep='first')
df_cheapest[['Company','Price_euros']]
```

```python
#most expensive
df_expensive=df_sorted.drop_duplicates(subset=['Company'],keep='last')
df_expensive[['Company','Price_euros']]
```

```python
# ARGUMENTS

#inplace:default False , when True make changes to original dataframe
#ignore_index=True: resulting axis will be labeled 0,1,..

df_expensive=df_sorted.drop_duplicates(subset=['Company'],keep='last',inplace=True,ignore_index=True)
df_sorted[['Company','Price_euros']]
```

# Rename Column
The rename() method in pandas is used to change the labels of rows (index) or columns in a DataFrame or Series.

**📌 Basic Syntax:**

DataFrame.rename(mapper=None, index=None, columns=None, inplace=False)

**🔑 Parameters:**

**columns:** Dictionary to rename column names

**index:** Dictionary to rename index labels

**inplace:**

True → modifies original DataFrame

False → returns a new DataFrame (default)

```python
# rename column and overwrite the values of the dataframe
df=df.rename(columns={'gender':'Gender'})
df
```

```python
#rename multiple columns and update the dataframe with the inplace argument
df.rename(columns={'race/ethnicity':'Race/Ethnicity','math score':'math score'.title()},inplace=True)
df
```

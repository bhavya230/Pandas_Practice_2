```python
import pandas as pd
```

```python
#reading the csv file
df=pd.read_csv("StudentsPerformance.csv")
df.head()
```

# 1 Math operations
## 1.1 Operations in Column

```python
#select a column and calculate total sum

#1 sum() built in function
print(sum(df['math score'])) # not ideal approach as would not understand DataFrame behaviour fully.
# its also slower for large dataset.

# another method - sum() method
print(df['math score'].sum())

# best choice- as handles missing values(NaN),Faster,More Features,desgined specially for dataFrames and series
```

```python
#count,mean,std,max,and min
print(df['math score'].count())
print(df['math score'].mean())
print(df['math score'].std())
print(df['math score'].min())
print(df['math score'].max())
print(type(df['math score'].count()))
```

```python
# easier method is : descibe()
print(df.describe())
print(type(df.describe()))
print(df.describe()['math score']) # accessing a particular column 
```

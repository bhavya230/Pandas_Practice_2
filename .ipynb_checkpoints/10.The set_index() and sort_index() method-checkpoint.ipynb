# 2 set_index()
The set_index() method in pandas is used to set one or more existing columns of a DataFrame as its index (row labels).

**📌 Basic Syntax**:

DataFrame.set_index(keys, drop=True, inplace=False)


**🔑 Parameters**:

**keys**: Column name(s) you want to set as the index.

**drop (default True)**: Removes the column(s) used as the new index

**inplace (default False)**: Modifies the original DataFrame if True

```python
# set new index as the row label (index) and modify the original dataframe
df.set_index('new index',inplace=True)
df
```

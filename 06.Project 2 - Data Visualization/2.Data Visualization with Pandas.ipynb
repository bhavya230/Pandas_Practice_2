```python
import pandas as pd
```

```python
#reading the csv file
df=pd.read_csv('population_total.csv')
```

```python
#first look at dataset
df
```

# 1 Making a Pivot Table
.pivot() : Returns reshaped dataframe organized by given index/column values("pivot without aggregation ")


```python
# dropping null values
df.dropna(inplace=True)
```

```python
# making pivot table
df_pivot=df.pivot(index='year',columns='country',values='population')
```

```python
#selectiong some countries
df_pivot=df_pivot[['United States','India','China','Indonesia','Brazil']]
```

```python
# showing pivot table
df_pivot
```

# 2 Lineplot

```python
df_pivot.plot(kind='line',xlabel='year',ylabel='population',title='population over the year')
```

# Barplot

## 3.1 Single Barplot

```python
#selecting only 1 year (2020)
df_pivot_2020=df_pivot[df_pivot.index.isin([2020])]
```

```python
#showing new dataframe
df_pivot_2020
```

```python
#Transform dataframe from rows to columns or vice-versa
df_pivot_2020=df_pivot_2020.T
df_pivot_2020
```
